```python
import numpy as np
from matplotlib import pyplot as plt 
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier, VotingClassifier, RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score, precision_recall_curve
from sklearn.neighbors import KNeighborsClassifier as Knn
from sklearn.linear_model import LogisticRegressionCV
from sklearn.tree import DecisionTreeClassifier
import warnings
import xgboost as xgb

warnings.filterwarnings("ignore", category=DeprecationWarning)

# We use these random seeds to ensure reproductibility
np.random.seed(1)
```

This little project can be found on the Kaggle platform at the following [link](https://www.kaggle.com/andrewmvd/fetal-health-classification). The challenge is to classify the health of a fetus as Normal, Suspect or Pathological based on cardiotocogram exam data. The dataset is composed of 1 csv file based on the research work of Ayres-de-Campos et al. [Ayres de Campos et al. (2000) SisPorto 2.0 A Program for Automated Analysis of Cardiotocograms. J Matern Fetal Med 5:311-318]. The exploratory data analysis has already been done and can be found at this [link](https://github.com/HocineLebbad/fetal-health-nn/blob/main/fetal-health_neural-network-approach.ipynb). The notebook at this link shows the performance of a neural network model for this classification task. 

On this notebook, we will try another approach:

- 1) Thanks to an extra-trees classifier, we will perform a feature selection
- 2) We will then compare three ensemble techniques : a voting classification algorithm (composed of a k-nearest neighbours classifier, a logistic regression algorithm and a classification tree), an extreme gradient boosting algorithm and a random forest classifier. 

Let's go !

```python
df = pd.read_csv('data/fetal_health.csv')

# Explanation for the removal of this feature on the link above
df = df.drop(['histogram_tendency'], axis=1)

X, y = df.drop(['fetal_health'],axis=1), df['fetal_health']

encoder = OneHotEncoder()
y_onehot = encoder.fit_transform(y.values.reshape(-1,1)).toarray()
```

Here, we use an extra-trees classifier to identify the most relevent features to keep to predict the fetal health.

```python
forest = ExtraTreesClassifier(n_estimators=200, random_state=42)

forest.fit(X, y)

importance = forest.feature_importances_

print(X.columns)
print(importance)
```

We can see that the feature importance varies a lot from one feature to the other. The goal of the following function is to define the least important features that we can drop before building the different models. 

```python
def features_to_drop(columns, importance, threshold):
    ''' 
    Input : 
    columns: columns that correspond to features
    importance: list that correspond to the importance of each feature
                (sum equal to 1)
    threshold : float < 1 is a parameter  
    
    Output : 
    col_sorted : all columns sorted by importance
    imp_sorted : all importances sorted from highest to lowest
    col_todrop : all columns we want to drop in further analysis 
    '''
    
    index = np.argsort(importance)[::-1]

    col_sorted, imp_sorted, col_todrop = [[], [], []]
    tmp = 0
    for i in range(np.size(importance)):
        col_sorted.append(columns[index[i]])
        imp_sorted.append(importance[index[i]])

        if tmp > threshold:
            col_todrop.append(columns[index[i]])
        
        tmp += importance[index[i]]

    return col_sorted, imp_sorted, col_todrop

_, _, col_todrop = features_to_drop(X.columns, importance, 0.90)

print('The least important features are:', col_todrop)

X = X.drop(col_todrop, axis=1)
```

```python
# Data standardization

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, 
                                                    stratify=y, random_state = 42)

X_train_oh, X_test_oh, y_train_oh, y_test_oh = train_test_split(X, y_onehot, test_size=0.30, 
                                               stratify=y_onehot, random_state = 42)

scaler1 = StandardScaler()
scaler2 = StandardScaler()

X_train = scaler1.fit_transform(X_train)
X_train_oh = scaler2.fit_transform(X_train_oh)

X_test = scaler1.transform(X_test)
X_test_oh = scaler2.transform(X_test_oh)
```

We define two functions to show the model performance.
- the output of **save_f1score()** is the f1-score (without any averaging) of the training and test sets
- the output of **save_confusion_matrix()** is the confusion matrix on unseen data (with some normalization)

```python
def save_f1score(model, X_train, y_train, y_test, y_pred, onehot=False):

    if onehot:
        training_score = f1_score(np.argmax(y_train, axis=1), 
                                  np.argmax(model.predict(X_train), axis=1), 
                                  average=None)
        
        test_score = f1_score(np.argmax(y_test,axis=1), 
                              np.argmax(y_pred,axis=1), 
                              average=None)
    else:
        training_score = f1_score(y_train, 
                                  model.predict(X_train), average=None)

        test_score = f1_score(y_test, y_pred, average=None)

    return training_score, test_score

def save_confusion_matrix(y_test, y_pred, onehot=False):

    if onehot:
        test_confusion = confusion_matrix(np.argmax(y_test, axis=1), 
                                          np.argmax(y_pred, axis=1))
    else:
        test_confusion = confusion_matrix(y_test, y_pred)

    conf_mat = []
    for i in range(3):
        conf_mat.append(test_confusion[:,i]/np.sum(test_confusion[:,i]))

    return conf_mat
```

Here, we build the three different models and we store their respective performance (f1-score and confusion matrix) to plot them afterwards.

```python
## Voting Classification

knn = Knn()
logreg = LogisticRegressionCV(random_state=42)
dtrees = DecisionTreeClassifier(random_state=42)

cl = [('LogReg', logreg),
      ('KNN', knn),
      ('Dec Trees', dtrees)]

vot_clas = VotingClassifier(estimators=cl, voting='soft')

vot_clas.fit(X_train, y_train)

y_pred_vot = vot_clas.predict(X_test)

conf_mat_voting = save_confusion_matrix(y_test, y_pred_vot, False)
f1_train_voting, f1_test_voting = save_f1score(vot_clas, X_train, y_train, y_test, y_pred_vot, False)

## Extreme Gradient Boosting

xgb_class = xgb.XGBClassifier(objective='multi:softmax', 
                              eval_metric='aucpr')

grad_boost = xgb_class.fit(X_train, y_train)

y_pred_gb = xgb_class.predict(X_test)

conf_mat_xgb = save_confusion_matrix(y_test, y_pred_gb, False)
f1_train_xgb, f1_test_xgb = save_f1score(grad_boost, X_train, y_train, y_test, y_pred_vot, False)

## Random Forest

rf = RandomForestClassifier(random_state=123)

rf.fit(X_train_oh, y_train_oh)

y_pred_rf = rf.predict(X_test_oh)

conf_mat_rf = save_confusion_matrix(y_test_oh, y_pred_rf, True)
f1_train_rf, f1_test_rf = save_f1score(rf, X_train_oh, y_train_oh, y_test_oh, y_pred_rf, True)
```

Now, we can see how our models perform.

```python
fig, ax = plt.subplots(1, 3, figsize=(18,5))

confusion_matrix_to_plot = [conf_mat_voting, conf_mat_xgb, conf_mat_rf]
titles = ['f1_train : %s \n f1_test: %s \n \n Voting classifier'%(f1_train_voting, f1_test_voting), 
          'f1_train : %s \n f1_test: %s \n \n Extreme gradient boosting'%(f1_train_xgb, f1_test_xgb), 
          'f1_train : %s \n f1_test: %s \n \n Random forest classifier'%(f1_train_rf, f1_test_rf)]

labels = ['Normal', 'Suspect', 'Pathological']

for i in range(3):
    sns.heatmap(confusion_matrix_to_plot[i], ax=ax[i], annot=True, fmt='.2%', cmap='Reds',
                xticklabels=labels, yticklabels=labels)
    ax[i].set_title(titles[i])
    ax[i].set_xlabel('\n Predicted')
    ax[i].set_ylabel('Actual')
```

It appears that the random forest classifier performs better than the neural network and the other ensemble techniques. Almost 94% of the 'Normal' fetus, 91% of the 'Suspect' fetus and 98% of the 'Pathological' fetus are correctly labeled!

The number of estimators of the random forest classifier is 10 by default (note: it has been changed to 100 recently according to the sklearn documentation). Let's see if the f1-score can be improved by increasing the number of estimators.

```python
f1test_class0, f1train_class0 = [[], []]
f1test_class1, f1train_class1 = [[], []]
f1test_class2, f1train_class2 = [[], []]

for i in range(1,100,5):
    rf = RandomForestClassifier(n_estimators=i)
    rf.fit(X_train_oh, y_train_oh)
    y_pred = rf.predict(X_test_oh)
    tmp1 = f1_score(y_test_oh, y_pred, average=None)
    tmp2 = f1_score(y_train_oh, rf.predict(X_train_oh), average=None)

    f1test_class0.append(tmp1[0])
    f1train_class0.append(tmp2[0])    
    f1test_class1.append(tmp1[1])
    f1train_class1.append(tmp2[1])    
    f1test_class2.append(tmp1[2])
    f1train_class2.append(tmp2[2])

f1test = [f1test_class0, f1test_class1, f1test_class2]
f1train = [f1train_class0, f1train_class1, f1train_class2]

fig, ax = plt.subplots(1, 3, figsize=(18,6))

for i in range(3):
    ax[i].plot(range(1, 100, 5),f1train[i], color='k', label='Train')
    ax[i].plot(range(1, 100, 5),f1test[i], color='r', label='Test')
    ax[i].set_xlabel('Number of estimators')
    ax[i].set_ylabel('f1-score')
    ax[i].legend()
```

The f1-scores on the train and test set increase when the number of estimators increases from 1 to ~10 and then reach a plateau. It does not seem appropriate to increase the number of estimators for this specific problem. Let's recap below the performances of the best developed model. We add the corresponding precision-recall curve.

```python
precision, recall = dict(), dict()

for i in range(3):
    precision[i], recall[i], _ = precision_recall_curve(y_test_oh[:,i], y_pred_rf[:,i])

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

colors = ['k','r','b']
labels = ['Normal', 'Suspect', 'Pathological']

sns.heatmap(conf_mat_rf, ax=ax[0], annot=True, fmt='.2%', cmap='Reds',
            xticklabels=labels, yticklabels=labels)

ax[0].set_title('f1_train : %s \n f1_test: %s '
                '\n \n Random forest classifier'%(f1_train_rf, f1_test_rf))
ax[0].set_xlabel('\n Predicted')
ax[0].set_ylabel('Actual')

for i in range(3):
    ax[1].plot(recall[i], precision[i], color=colors[i], label=labels[i], linestyle='-')

ax[1].set_xlabel('Recall TP/(FN+TP)')
ax[1].set_ylabel('Precision TP/(FP+TP)')

plt.legend()
plt.show()
```

As suggested by the confusion matrix that corresponds to the random forest classifier model, the area under the precision-recall curve for the 'Healthy' and 'Pathological' classes is higher than for the 'Suspect' class!
